```python
import shutil
import os
```

```python
dir_out = 'All'
if not os.path.exists(dir_out):
    os.makedirs(dir_out)

dir_all = os.listdir('./')
count = 1
for dir_name in dir_all:
    if not os.path.isdir(dir_name):
        continue
    
    filename_imgs = os.listdir(dir_name)
    for filename_img in filename_imgs:
        full_path = os.path.join(dir_name, filename_img)
        full_path_out = os.path.join(dir_out, 'img'+'%03d'%count+'.png')
        print('From', full_path, 'To', full_path_out)
        count = count+1
        shutil.copy(full_path, full_path_out)
```
